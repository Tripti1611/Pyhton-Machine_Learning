# Guess the Word Game

In this game, there is a list of words present, out of which our interpreter will choose 1 random word. The user first has to input their names and then, will be asked to guess any alphabet. If the random word contains that alphabet, it will be shown as the output(with correct placement) else the program will ask you to guess another alphabet. User will be given 12 turns(can be changed accordingly) to guess the complete word.

```python
import random

print("Welcome to 'Guess The Word' Game!")
name = input("Enter your name: ")
print("Good Luck {0}!".format(name))

'''Fetching words from file and converting it in list'''
f = open("GuessGame.txt")
words = f.read().split()

'''Choosing one word from the list of words'''
word = random.choice(words)

print("Guess the charcaters!!!!")
guesses = '' #it will display characters guessed

'''total turns'''
turns = 12

while turns > 0:
	'''no of times user failed can be counted'''
	failed = 0

	#printing all characters of input word one at a time
	for char in word:
		#comparing char with char in guesses
		if char in guesses:
			print(char)
		else:
			print('-')
			failed += 1 #for every failure increment failed variable

	if failed == 0:
		#user won the game
		print("Congrats! You Won!!")
		print("The word is: '{0}'".format(word))
		break

	#if user guessed wrong char ask user to enter another char
	guess = input("Guess a character: ")

	#check a single char is entered or not
	if len(guess) == 1:
		#every input char will be stored in guesses
		guesses += guess

		if guess not in word:
			#wrong guess decrement turns
			print("Wrong Guess!")
			turns -= 1
			print("You have {0} more turns".format(turns))
	else:
		print("Enter single character only!!")

	if turns == 0:
		print("You Loose! Try Again!!")
```
