# Plotting 3D Graphs

To plot 3D graphs using Matplotlib we have to import mplot3d module from mpl_toolkits that is included in Matplotlin installation.

```python
#viewing figure of 3D graph
from mpl_toolkits import mplot3d

import numpy as np
import matplotlib.pyplot as py
fig = py.figure()
ax = py.axes(projection="3d")
```

```python
#adding lines in 3D graph
from mpl_toolkits import mplot3d

import numpy as np
import matplotlib.pyplot as py
fig = py.figure()
ax = py.axes(projection="3d")
#data for 3d line
zline = np.linspace(0,15,1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline,yline,zline,'gray')
```

```python
#adding scatter points in 3D graph
from mpl_toolkits import mplot3d

import numpy as np
import matplotlib.pyplot as py
fig = py.figure()
ax = py.axes(projection="3d")
#data for 3D scatter points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata)+0.1 * np.random.randn(100)
ydata = np.cos(zdata)+0.1 * np.random.randn(100)
ax.scatter3D(xdata,ydata,zdata, c=zdata, cmap='Greens')
```

```python
#adding lines and points on 3D graph
from mpl_toolkits import mplot3d

import numpy as np
import matplotlib.pyplot as py
fig = py.figure()
ax = py.axes(projection="3d")
#data for 3d line
zline = np.linspace(0,15,1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline,yline,zline,'purple')
#data for 3D scatter points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata)+0.5 * np.random.randn(100)
ydata = np.cos(zdata)+0.1 * np.random.randn(100)
ax.scatter3D(xdata,ydata,zdata,c=zdata,cmap=None)
```

# 3D Contour Plots

```python
#3D Contour Plots
from mpl_toolkits import mplot3d

%matplotlib inline
import numpy as np
import matplotlib.pyplot as py

def f(x,y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-6,6,20)
y = np.linspace(-6,6,20)

X,Y = np.meshgrid(x,y)
Z = f(X,Y)
fig = py.figure()
ax = py.axes(projection='3d')
ax.contour3D(X,Y,Z,50,cmap='binary')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

# Surface Triangulation

```python
#Surface Triangulation
from mpl_toolkits import mplot3d

%matplotlib inline
import numpy as np
import matplotlib.pyplot as py

theta = 2*np.pi*np.random.random(100)
r = 6*np.random.random(100)
x = np.ravel(r * np.sin(theta))
y = np.ravel(r * np.cos(theta))
z = f(x,y)
ax = py.axes(projection='3d')
ax.plot_trisurf(x,y,z,cmap='Greens')
ax.scatter(x,y,z,c=z,cmap='viridis')
```
