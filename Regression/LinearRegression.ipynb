# Linear Regression

Using Linear Regression to predict net worth based on person's age.

Function is used for training the model

```python
#Importing necessary libraries
import numpy
import random
import matplotlib.pyplot as plt

#defining function for training the model
def student_Reg(ages_train,net_worth_train):
	from sklearn.linear_model import LinearRegression
	reg = LinearRegression()                     #creating the model
	reg.fit(ages_train, net_worth_train)         #fitting/training the model
	return reg

#Generating dataset with random values
random.seed(42)
numpy.random.seed(42)

ages=[]
for i in range(100):
	ages.append(random.randint(20,80))
net_worths = [i*6.25 + numpy.random.normal(scale=40.) for i in ages] #this gives equation of linear regression

#Reshaping the generated dataset
ages = numpy.reshape(numpy.array(ages), (len(ages),1))
net_worths = numpy.reshape(numpy.array(net_worths), (len(ages),1))

#Splitting dataset to obtain training and testing data
from sklearn.model_selection import train_test_split
ages_train, ages_test, net_worths_train, net_worths_test = train_test_split(ages, net_worths)

#Calling the function for training the model
reg = student_Reg(ages_train,net_worths_train)

#Printing slope and intercept
print("Coefficient/Slope = ",reg.coef_)
print("Intercept = ",reg.intercept_)

#Score of data
print("R-squared score of Training data ",reg.score(ages_train,net_worths_train))
print("R-squared score of Testing data ",reg.score(ages_test,net_worths_test))

#Accuracy of the model
from sklearn.metrics import r2_score
print("R-square accuracy of model: ",r2_score(net_worths_test,reg.predict(ages_test))) #predicting data
```

Plotting the training, testing and predicted data in graph

```python
plt.figure(figsize=(10,8))
plt.scatter(ages_train,net_worths_train,color="green",label="training data") #plotting training data
plt.scatter(ages_test,net_worths_test,color="red",label="testing data")    #plotting testing data with actual values
plt.plot(ages_test,reg.predict(ages_test),color="yellow",marker="o",label="predicted data")     #plotting testing data with predicted values
plt.xlabel("Ages")
plt.ylabel("Net Worth")
plt.legend(loc=2)
plt.show()
```

```python
#Importing necessary libraries
import numpy
import random
import matplotlib.pyplot as plt

#Creating dataset
random.seed(42)
numpy.random.seed(42) #used to give same data everytime

ages = []
for i in range(100):
    ages.append(random.randint(20,85))
    
net_worth = [i*6.25 + numpy.random.normal(scale=40) for i in ages] #gives linear equation for regression

#Reshaping generated dataset
ages = numpy.reshape(numpy.array(ages),(len(ages),1))
net_worth = numpy.reshape(numpy.array(net_worth),(len(ages),1))

#Splitting dataset to obtain training and testing data
from sklearn.model_selection import train_test_split
ages_train,ages_test,net_worth_train,net_worth_test = train_test_split(ages,net_worth)

#Importing Libraries for Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
reg = LinearRegression()             #Creating model
reg.fit(ages_train,net_worth_train)  #Training model

#Dsiplay slope and intercept
print("Slope: ",reg.coef_)
print("Intercept: ",reg.intercept_)

#Display score of data
print("Train Score: ",reg.score(ages_train,net_worth_train))
print("Test Score: ",reg.score(ages_test,net_worth_test))

#Checking accuracy of the model
print("R-square Accuracy of model: ",r2_score(net_worth_test,reg.predict(ages_test))) #predict() used for predicting
```

Plotting training, testing and predicted data in graph

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
plt.scatter(ages_train,net_worth_train,color="purple",label="training data")
plt.scatter(ages_test,net_worth_test,color="cyan",label="testing data")
plt.plot(ages_test,reg.predict(ages_test),color="magenta",marker="o",label="predicted data")
plt.xlabel("Ages")
plt.ylabel("Net Worth")
plt.legend(loc=4)
plt.show()
```
